```python
import numpy as np

# 假設多維數據
data = np.random.random((100, 120, 2))  # 3D 數據

# 多維傅立葉變換
fft_results = np.fft.fftn(data)

# 提取頻率
frequencies = [np.fft.fftfreq(n) for n in data.shape]
```

```python
from scipy.signal import csd
# 計算交叉譜
frequencies, cross_spectrum = csd(data[:, :, 0], data[:, :, 1], fs=1000, nperseg=120)
print(frequencies.shape, cross_spectrum.shape)
```

```python
print("Frequencies:", frequencies)
print("Cross Spectrum:", cross_spectrum)
```

```python
print("FFT Results Shape:", fft_results.shape)
print("Frequencies:", frequencies)
```
