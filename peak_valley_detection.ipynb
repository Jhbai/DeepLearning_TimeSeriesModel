```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs

# 生成模擬數據
data, _ = make_blobs(n_samples=300, centers=1, cluster_std=0.60, random_state=42)
# 加入一些異常點
outliers = np.random.uniform(low=-6, high=6, size=(20, 2))
data = np.vstack([data, outliers])

# 建立 Isolation Forest 模型
model = IsolationForest(n_estimators=100, contamination=0.15, random_state=42)
model.fit(data)

# 預測異常
scores_pred = model.decision_function(data)
is_anomaly = model.predict(data)

# 可視化結果
plt.scatter(data[:, 0], data[:, 1], c=is_anomaly, cmap='Paired')
plt.title('Isolation Forest Anomaly Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.show()
```

```python
arr = np.random.normal(0, 1, size = (1000, 1))
arr[100] = 10

model = IsolationForest(n_estimators=200, contamination=0.01, max_samples='auto', random_state=42)
model.fit(arr)
is_anomaly = model.predict(arr)
is_anomaly = np.where(is_anomaly==-1)[0]

plt.figure(figsize = (6, 2))
plt.plot(arr, color = "gray")
plt.scatter(is_anomaly, arr[is_anomaly], color = "red")
plt.grid()
plt.show()
```
